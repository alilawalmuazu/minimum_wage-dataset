The results of the linear regression analysis provide valuable insights into how the Dollar Rate and Inflation affect the Minimum Wage in Naira. Let's break down the meaning of the coefficients, intercept, and R-squared value:

### Coefficients
The coefficients represent the change in the dependent variable (Minimum Wage) for a one-unit change in each independent variable, holding all other variables constant.

- **Dollar Rate (Naira) Coefficient: 79.80306911**
  - This coefficient indicates that for every 1 Naira increase in the Dollar Rate, the Minimum Wage increases by approximately 79.80 Naira, assuming inflation remains constant.
  - This positive coefficient suggests that as the Naira depreciates against the Dollar (i.e., the Dollar Rate increases), the minimum wage tends to increase. This could be because a higher exchange rate might lead to higher costs of imported goods, putting pressure on wages to rise to maintain purchasing power.

- **Inflation (%) Coefficient: -521.76483576**
  - This coefficient indicates that for every 1% increase in inflation, the Minimum Wage decreases by approximately 521.76 Naira, assuming the Dollar Rate remains constant.
  - This negative coefficient suggests that higher inflation is associated with a decrease in the real value of the minimum wage. It implies that inflation erodes the purchasing power of the wage, and unless adjusted, the nominal wage might not keep up with rising prices, effectively reducing the wage's real value.

### Intercept
- **Intercept: 5203.335817258287**
  - The intercept represents the estimated Minimum Wage when both the Dollar Rate and Inflation are zero. While this scenario is hypothetical and unrealistic, the intercept serves as a baseline value for the model.
  - In practical terms, it provides a starting point for understanding how the combination of Dollar Rate and Inflation affects the Minimum Wage.

### R-squared
- **R-squared: 0.8646595487826099**
  - The R-squared value measures the proportion of variance in the dependent variable (Minimum Wage) that is explained by the independent variables (Dollar Rate and Inflation) in the model.
  - An R-squared value of approximately 0.865 indicates that 86.5% of the variability in the Minimum Wage can be explained by changes in the Dollar Rate and Inflation.
  - This relatively high R-squared value suggests that the model fits the data well and that the Dollar Rate and Inflation are significant predictors of the Minimum Wage.

### Interpretation
Overall, the analysis suggests that both the Dollar Rate and Inflation have significant impacts on the Minimum Wage in Nigeria:
- An increase in the Dollar Rate (a depreciation of the Naira) is associated with an increase in the Minimum Wage, likely due to the need to adjust wages to keep up with the cost of living.
- Higher inflation is associated with a decrease in the real value of the Minimum Wage, highlighting the need for regular adjustments to ensure that wages maintain their purchasing power over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score



# Create DataFrame
df = pd.DataFrame(data)

# Plot the trends
plt.figure(figsize=(14, 7))

plt.subplot(2, 2, 1)
plt.plot(df['Year'], df['Minimum Wage (Naira)'], marker='o')
plt.title('Minimum Wage over the Years')
plt.xlabel('Year')
plt.ylabel('Minimum Wage (Naira)')

plt.subplot(2, 2, 2)
plt.plot(df['Year'], df['Dollar Rate (Naira)'], marker='o', color='green')
plt.title('Dollar Rate over the Years')
plt.xlabel('Year')
plt.ylabel('Dollar Rate (Naira)')

plt.subplot(2, 2, 3)
plt.plot(df['Year'], df['Inflation (%)'], marker='o', color='red')
plt.title('Inflation over the Years')
plt.xlabel('Year')
plt.ylabel('Inflation (%)')

plt.subplot(2, 2, 4)
sns.scatterplot(x='Dollar Rate (Naira)', y='Minimum Wage (Naira)', data=df, color='purple', label='Dollar Rate')
sns.scatterplot(x='Inflation (%)', y='Minimum Wage (Naira)', data=df, color='orange', label='Inflation')
plt.title('Minimum Wage vs Dollar Rate and Inflation')
plt.xlabel('Dollar Rate (Naira) / Inflation (%)')
plt.ylabel('Minimum Wage (Naira)')
plt.legend()

plt.tight_layout()
plt.show()

# Prepare data for regression
X = df[['Dollar Rate (Naira)', 'Inflation (%)']]
y = df['Minimum Wage (Naira)']

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Calculate R-squared
r2 = r2_score(y, y_pred)

# Display coefficients and R-squared
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("R-squared:", r2)
```
